```python
from youtube_transcript_api import YouTubeTranscriptApi
import urllib
```

```python
ytt_api = YouTubeTranscriptApi()
```

```python
url_1 = "https://www.youtube.com/watch?v=xKZ_8ULR91Y"
url_2 = "https://youtu.be/xKZ_8ULR91Y"

parsed_1 = urllib.parse.urlparse(url_1)
print(f"Location 1: {parsed_1.netloc}")
print(f"Path 1: {parsed_1.path}")
print(f"Query 1: {parsed_1.query}")

parsed_2 = urllib.parse.urlparse(url_2)
print(f"Location 2: {parsed_2.netloc}")
print(f"Path 2: {parsed_2.path}")
print(f"Query 2: {parsed_2.query}")
```

```python
def extract_video_id(url: str) -> str:
    parsed_url = urllib.parse.urlparse(url)
    
    if parsed_url.netloc == "www.youtube.com":
        query_param = parsed_url.query
        video_id = query_param.split("=")[1]
        return video_id
    
    elif parsed_url.netloc == "youtu.be":
        query_path = parsed_url.path
        video_id = query_path.split("/")[1]
        return video_id
        
    else:
        msg = (
            "Do not recognize YouTube url."
            "Make sure it's of format `https://www.youtube.com/watch/v=[video_id]"
            "or `https://youtu.be./[video_id]"
        )
        raise ValueError("Do not recognize YouTube url. Make sure it's of format `www.youtube.com/watch/v=[video_id]")
```

```python
video_id = extract_video_id(url = url_1)
```

```python
response = ytt_api.fetch(video_id)
```
